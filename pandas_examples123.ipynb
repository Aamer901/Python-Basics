```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
plt.figure(figsize=(4,4))
plt.hist(df1['Accept'])
plt.xlabel('Accept')
plt.ylabel('Count')
plt.title('histogram of accept')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Count')
plt.title('histogram of SAT')
plt.show()
```
